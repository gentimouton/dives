```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

# https://github.com/veekun/pokedex/blob/master/pokedex/data/csv/experience.csv
df = pd.read_csv('data/experience.csv')
df.head()
```

```python
palette = sns.color_palette("deep", df['growth_rate_id'].nunique())
ax = sns.lineplot(x="level", y="experience", hue="growth_rate_id", data=df.query('experience>0 & level<50'), palette=palette)
# ax.set(yscale="log", xscale='log')
```
