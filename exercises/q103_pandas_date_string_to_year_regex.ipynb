Python Pandas Data Manipulation External Dataset

Given [this dataset](https://drive.google.com/file/d/1f_mVrLYG5quI38L1tA2MQwmyppCJd2Fb/view), 
write code to put the Cease Date column into YYYY format. 
You'll notice the column is currently in a mix of various formats. 
We're going to use the year in the Cease Date column in a later related question.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as spstats
import re

df = pd.read_csv('q103_data.csv')
df.head()
```

```python
# possible date formats in file: 05/2012, 2012, Not Stated

df['cease_date_clean'] = (
    df['Cease Date']
    .map(
        lambda s: re.findall('(\d{4})', s) or np.nan,
        na_action='ignore'
    )
)
df[['Cease Date', 'cease_date_clean']].sample(20)
```
