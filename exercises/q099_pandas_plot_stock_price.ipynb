Using Python, create a line chart with both the daily and 7-day moving average for the close price for Apple stock. You can download a CSV with prices [here](https://finance.yahoo.com/quote/AAPL/history/).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as spstats

# Date,Open,High,Low,Close,Adj Close,Volume
df = pd.read_csv('q099_data.csv', parse_dates=['Date'])
df.head()
```

```python
df = df.set_index('Date')
df['Close_mavg7'] = df['Close'].rolling(window=7).mean()
df[['Close','Close_mavg7']].plot()
```
