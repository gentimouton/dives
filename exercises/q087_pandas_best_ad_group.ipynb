# Question 87 - Best ad group

Given the following [data set](https://docs.google.com/spreadsheets/d/1qY-OMOX59hc72KAmwsIa3xnPr2rXCyPnK9cE7OjPc2U/edit#gid=0), can you identify the top 3 performing ad groups? Explain how you evaulated the ad groups.

The solution for premium users will be written using Python Pandas. Note there can be several correct solutions for this problem.



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('q087_data.csv')
df.head()
```

```python
# Option 1: best ad group makes the most benefit per impression
df['cost'] = df['avg_cost_per_click'] * df['clicked']
agg = df.groupby('ad').agg({'total_revenue':sum, 'shown':sum, 'clicked':sum, 'converted':sum, 'cost':sum})
agg['benefit_per_imp'] = (agg['total_revenue'] - agg['cost']) / agg['shown']
agg.sort_values(by='benefit_per_imp', ascending=False).head(3)
```

```python
# Option 2: best ad group converts the best
agg['conv'] = agg['converted'] / agg['shown'] # could also be clicked/shown
agg.sort_values(by='conv', ascending=False).head(3)
```

# conclusion

Groups 16,2,14 generate the most benefit (revenue minus cost).
Groups 13,34,18 convert the best, ie highest p(purchase | shown).
