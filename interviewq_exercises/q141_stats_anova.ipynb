```python
# use ANOVA when comparing means between 3+ groups
# its null hypothesis is: mu1 == mu2 == mu3
# if ANOVA returns p<.05, it means 1+ group differs from at least 1+ other.
# to know which group(s) differ, run post-hoc t-tests.

# https://pythonfordatascience.org/anova-python/
import scipy.stats as stats
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
df.head()
```

```python
fares_for_each_class = [
    df[df['class']==cls]['fare'] for cls in df['class'].unique()
]
f, p = stats.f_oneway(*fares_for_each_class)
print(f"f={f:.2f}, p={p:.3f}")

df.groupby('class').agg({'fare':['mean','std','count']})
```
