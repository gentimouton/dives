# Plotting daily earnings (given balances)
Data Analysis Python Pandas Data Manipulation Data Visualization External Dataset

Given the following [dataset](https://docs.google.com/spreadsheets/d/1FRt2JwUPU_1bruGejUg0vAa_IHtFBn5VdCSN-kkTdpM/edit#gid=0) of dates and end of day balances, can you plot daily earnings in a chart?

```python
import pandas as pd
import matplotlib.pyplot as plt

filename = 'q139_data_daily_earnings.csv'
df = pd.read_csv(filename) #, encoding='utf-8') # pesky \xa0
df.head()
```

```python
df['earnings'] = df['Balance'] - df['Balance'].shift()
df['earnings_mavg31'] = df['earnings'].rolling(31).mean()
df.set_index('Date')[['earnings','earnings_mavg31']].plot()
```
